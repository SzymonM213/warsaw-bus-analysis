```python
from visualization.overspeed import count_overspeeding_vehicles_from_hour
```

```python
overspeeding_vehicles_17, street_to_overspeeds_17 = count_overspeeding_vehicles_from_hour(17)
```

```python
overspeeding_vehicles_2, street_to_overspeeds_2 = count_overspeeding_vehicles_from_hour(2)
```

```python
import matplotlib.pyplot as plt

x = ['2-3', '17-18']
y = [overspeeding_vehicles_2, overspeeding_vehicles_17]

plt.bar(x, y)
for i in range(len(x)):
    plt.text(i, y[i], str(y[i]), ha='center', va='bottom')
plt.title('Overspeeding vehicles')
plt.xlabel('Hour')
plt.show()
```

```python
overspeeding_vehicles_17_count = {k: len(v) for k, v in street_to_overspeeds_17.items()}
overspeeding_vehicles_2_count = {k: len(v) for k, v in street_to_overspeeds_2.items()}
```

```python
from typing import Dict

def plot_overspeeds(overspeeds: Dict[str, int], title: str, n: int=10):
    streets = list(overspeeds.keys())[:n]
    counts = list(overspeeds.values())[:n]

    streets = [str(i) for i in streets]

    _, ax = plt.subplots()
    ax.barh(streets, counts)
    ax.set_xlabel('Number of overspeeding vehicles')
    ax.set_title(title)

    plt.show()
```

```python
plot_overspeeds(overspeeding_vehicles_17_count, 'Overspeeding vehicles per street between 17 and 18 on most overspeeding streets')
```

```python
plot_overspeeds(overspeeding_vehicles_2_count, 'Overspeeding vehicles per street between 2 and 3 on most overspeeding streets')
```
