```python
from visualization.punctuality import get_delays_from_hour, filter_delays
import matplotlib.pyplot as plt
```

```python
delays_17 = get_delays_from_hour(17)
delays_17_filtered = filter_delays(delays_17, 5)
```

```python
delays_5 = get_delays_from_hour(5)
delays_5_filtered = filter_delays(delays_5, 5)
```

```python
delays_17_number = len(delays_17_filtered)
delays_5_number = len(delays_5_filtered)

x = ['5-6', '17-18']
y = [delays_5_number, delays_17_number]

plt.bar(x, y)
for i in range(len(x)):
    plt.text(i, y[i], str(y[i]), ha='center', va='bottom')

plt.title('Buses delayed more than 5 minutes')
plt.xlabel('Hour')
plt.ylabel('Number of delayed buses')
plt.show()
```

```python
avg_delay_17 = delays_17['Delay'].mean()
avg_delay_2 = delays_5['Delay'].mean()

x = ['5-6', '17-18']
y = [avg_delay_2, avg_delay_17]

plt.bar(x, y)
for i in range(len(x)):
    plt.text(i, y[i], str(round(y[i], 2)), ha='center', va='bottom')

plt.title('Average delay')
plt.xlabel('Hour')
plt.ylabel('Average delay in minutes')
plt.show()
```

```python
delays_17_filtered['Delay'].plot.kde(c='r' , label='17-18')
delays_5_filtered['Delay'].plot.kde(c='b', label='5-6')
plt.legend()
plt.title('Delay distribution')
plt.xlabel('Delay in minutes')
plt.ylabel('Density')
plt.show()
```
